```python

from matplotlib import pyplot as plt
from simulation import Simulation
import random
import numpy as np
from gym.utils import seeding


configs = ['sa','caa','coa','sa_dro','caa_dro','coa_dro']

def add_metrics(x,y):
    res = {}
    for key in x.keys():
        res[key] = []
        for idx in range(min(len(x[key]),len(y[key]))):
            res[key].append(x[key][idx]+y[key][idx])
        
    return res

instances = {}
results = {}
N=10
for config in configs:
    for idx in range(N):
        random.seed(idx+100)
        _, _ = seeding.np_random(idx+100)
        np.random.seed(idx+100)
        instances[config] = Simulation(config=f'./configs/{config}.json')
        print(f'Running simulation with config {config} -- {idx}')
        instances[config].run()
        if idx==0:
            results[config] = instances[config].get_metrics()
        else:
            results[config] = add_metrics(results[config],instances[config].get_metrics())
    for key in results[config].keys():
        for idx in range(len(results[config][key])):
            results[config][key][idx]/=N
    with open(f'{config}.txt', "w") as f:
        str_dictionary = repr(results[config])
        f.write("results = " + str_dictionary + "\n")
```

```python
def plot_box(metric):
    data = [results[config][metric] for config in configs]
    fig, ax = plt.subplots()
    # build a violin plot
    ax.boxplot(data, showmeans=True)
    # add title and axis labels
    ax.set_xlabel('timeslot (s)')
    ax.set_ylabel(metric)
    # add x-tick labels
    xticklabels = configs
    ax.set_xticklabels(xticklabels)
    # add horizontal grid lines
    ax.yaxis.grid(True)
    # show the plot
    plt.show()
```

```python
plot_box('service_capability')
```
